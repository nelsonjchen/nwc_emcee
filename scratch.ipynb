```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('sheets/mario.png')

RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
HSV_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
GRAY_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(RGB_img, cmap='Greys_r')
plt.title('my picture')
plt.show()


lower_white = np.array([0, 0, 0])
upper_white = np.array([85, 255, 255])

_, masked= cv2.threshold(GRAY_img, 10, 255, cv2.THRESH_BINARY)
masked = cv2.bitwise_not(masked)

plt.imshow(masked, cmap="Greys")
plt.title('greens')
plt.show()
```

```python
palette_x = 128
pallete_y = 7
digit_images = []
# Digit Images
for digit in range(10):
    origin_palette_x = palette_x + 8 * digit
    digit_images.append(
        masked[0:7, origin_palette_x:origin_palette_x+8]
    )
for i, digit_image in enumerate(digit_images):
    plt.imshow(digit_image, cmap="Greys")
    plt.title(f"{i}")
    plt.show()

```
